```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
def Plot_Covid_19_Cases(Countries_list, date_min, date_max):
    url = "https://covid.ourworldindata.org/data/ecdc/total_cases.csv"
    df = pd.read_csv(url, index_col='date', parse_dates=True)
    ax = plt.gca()
    for c in Countries_list:
        if not c in df.columns:
            print(c, " - not found")
            continue
        ss = df.plot(kind ='line',  y = c, ax = ax, figsize = (10,10))
    ax.set_ylabel("Number of cases")
    ax.set_xlim([date_min, date_max])
    plt.show()
    return 
```

```python
c_list = ["Oman", "Bahrain", "United Arab Emirates"]
Plot_Covid_19_Cases(c_list, '2020-4-5', '2020-6-10')
```
